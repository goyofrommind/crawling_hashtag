```python
from wordcloud import WordCloud              # 워드클라우드 패키지
import matplotlib.pyplot as plt              #  데이터를 시각화
from collections import Counter              #  텍스트 및 횟수 추출
from konlpy.tag import Okt                   #  한국어를 처리하는 형태소 분석 패키지 
from konlpy.tag import Kkma                  #  형태소를 조금씩 다르게 처리한다  명사 ,동사 등등
from wordcloud import WordCloud, STOPWORDS   # 예외단어 처리용도 
from PIL import Image                        #  그림을 불러오는 패키지
import numpy as np                           #  그림을 배열로 나타내는 패키지
```

```python
with open('E:/ts/벚꽃.csv', 'r', encoding='utf-8') as f:
    text = f.read()
okt = Okt()
nouns = okt.nouns(text) # 명사만 추출

words = [n for n in nouns if len(n) > 1] # 단어의 길이가 1개인 것은 제외

count = Counter(words) # 위에서 얻은 words를 처리하여 단어별 빈도수 형태의 딕셔너리 데이터를 구함
# count = dict(count.most_common(n=20)) ## 상위20개 
```

```python
wordcloud = WordCloud(font_path='malgun', width=400, height=400, scale=2.0, max_font_size=250)
wc =wordcloud.generate_from_frequencies(count)
plt.figure()
plt.imshow(wc)
```
